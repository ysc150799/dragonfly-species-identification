# Dragonfly Data Segregation and Cleaning

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datefinder
```

## Read and Clean Data

```python
data=pd.read_csv("data_600.csv")
data["feed__data__message"]=data["feed__data__message"].str.lower()
data.drop(["feed__data__id","feed__paging__previous","feed__paging__next","id"],axis=1,inplace=True)
data.head(15)
```

```python
data["species_name"]=data["feed__data__message"].str.find("vestalis gracilis")
```

```python
data.head()
```

## Read Species list and convert to Lower case

```python
species=pd.read_csv("Dragonfly_species.csv")
species["Aciagrion"]=species["Aciagrion"].str.lower()
species["Aciagrion"]=species["Aciagrion"].str.split()
species.head(15)
```

## Clean species Data

```python
s = " "
sk = []
for i in range(0,699):
    spe = s.join((species["Aciagrion"][i])[:2])
    sk.append(spe)   
```

## Compare and match species name 

```python
for i in range(0,699):
    data[sk[i]]=data["feed__data__message"].str.find(sk[i])
```

```python
datax=data.iloc[:,7:-1]
listt=datax.columns
listt[1]
datax.head()
```

## Add Species name to the dataset

```python

for i in range(0,600):
    for j in listt:
        if datax[j].iloc[i] >= 0:
            data["species_name"].iloc[i]=j
```
