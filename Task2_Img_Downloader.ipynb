## Create Directories

```python
import os 
parent_dir = "/home/yashd/Documents/dragonfly-species-identification/Img/"
for row in df1.itertuples():
    Nolabelpath="/home/yashd/Documents/dragonfly-species-identification/Img/Nolabel"
    Nolabelexists=os.path.exists(Nolabelpath)
    speciespath=os.path.join(parent_dir,getattr(row,"species_name"))
    speciesexists=os.path.exists(speciespath)
    if getattr(row,"species_name") == "-1" and Nolabelexists == False:
        dir1="Nolabel"
    elif getattr(row,"species_name") != "-1" and speciesexists == False:
        dir1 = getattr(row,"species_name")
    elif Nolabelexists == True:
        continue
    path = os.path.join(parent_dir, dir1)
    os.mkdir(path)
    
```
