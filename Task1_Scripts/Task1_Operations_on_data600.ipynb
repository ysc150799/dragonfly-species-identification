# Dragonfly Data Segregation and Cleaning

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#import datefinder
#from dateutil import parser
from date_extractor import extract_dates
#from nltk.tag import StanfordNERTagger
#from cliff.api import Cliff
#from geotext import GeoText
import geograpy3
from googlegeocoder import GoogleGeocoder
from geopy.geocoders import Nominatim
import geocoder
```

## Read and Clean Data

```python
data=pd.read_csv("data_600.csv")
fdata=pd.read_csv("data_600.csv")
tpdata=pd.read_csv("data_600.csv")
data["feed__data__message"]=data["feed__data__message"].str.lower()
tpdata["feed__data__message"]=data["feed__data__message"].str.lower()
data.drop(["feed__data__id","feed__paging__previous","feed__paging__next","id"],axis=1,inplace=True)
fdata.drop(["feed__data__id","feed__paging__previous","feed__paging__next","id"],axis=1,inplace=True)
tpdata.drop(["feed__data__id","feed__paging__previous","feed__paging__next","id"],axis=1,inplace=True)
data.dropna(inplace=True)
fdata.dropna(inplace=True)
tpdata.dropna(inplace=True)
print(data.describe())
data.head(15)
```

```python
data["species_name"]=data["feed__data__message"].str.find("vestalis gracilis")
fdata["species_name"]=fdata["feed__data__message"].str.find("vestalis gracilis")
```

```python
data.head()
```

## Read Species list and convert to Lower case

```python
species=pd.read_csv("Dragonfly_species.csv")
species["Aciagrion"]=species["Aciagrion"].str.lower()
species["Aciagrion"]=species["Aciagrion"].str.split()
species.head(15)
```

## Add common name list

```python
common_name=pd.read_csv("Common_names.csv")
common_name.dropna(inplace=True)
common_name.to_csv("cname.csv")
tpdata.head()
```

## Clean species Data

```python
s = " "
sk = []
for i in range(0,699):
    spe = s.join((species["Aciagrion"][i])[:2])
    sk.append(spe)   
```

## Compare and match species name 

```python
for i in range(0,699):
    data[sk[i]]=data["feed__data__message"].str.find(sk[i])
```

```python
datax=data.iloc[:,7:-1]
listt=datax.columns
#listt[1]
datax.head()
```

## Add Species name to the dataset

```python
print(listt)
for i in range(0,584):
    for j in listt:
        if datax[j].iloc[i] >= 0:
            #print(j)
            fdata["species_name"].iloc[i]=j
```

```python
for i in range(0,584):
    fdata["species_name"].iloc[i]=str(fdata["species_name"].iloc[i])
print(fdata["species_name"].iloc[0])
for i in range(0,584):
    #print(fdata["species_name"].iloc[i])
    tpdata["feed__data__message"].iloc[i]=tpdata["feed__data__message"].iloc[i].replace(fdata["species_name"].iloc[i],' ')
```

## Process Common Names

```python
cname=pd.read_csv("cname.csv")
cname["COMMON NAME"]=cname["COMMON NAME"].str.lower()
cname.head()
```

```python
s = " "
sk1 = []
for i in range(0,108):
    spe = (cname["COMMON NAME"][i])
    sk1.append(spe)
sk1
```

```python
for i in range(0,108):
    tpdata[sk1[i]]=tpdata["feed__data__message"].str.find(sk1[i])
```

```python
datay=tpdata.iloc[:,2:]
listt1=datay.columns
datay.head()
```

```python
for i in range(0,584):
    for j in listt1:
        if datay[j].iloc[i] >= 0:
            tpdata["common blue jewel"].iloc[i]=j
```

```python
tpdata=tpdata.iloc[:,:3]
tpdata.head()
fdata["Common Name"]=tpdata["common blue jewel"]
fdata.head()
```

## Remove Delimiters

```python
for i in range(0,584):
    #print(data["feed__data__message"].iloc[i])
    #data["feed__data__message"].iloc[i]=data["feed__data__message"].iloc[i].split("\n")
    #data["feed__data__message"].iloc[i]=" ".join(data["feed__data__message"].iloc[i])
    #data["feed__data__message"].iloc[i]=data["feed__data__message"].iloc[i].replace('\r',' ')
    fdata["feed__data__message"].iloc[i]=fdata["feed__data__message"].iloc[i].replace('\\n',' ')
fdata.head()

```

## Finding Dates

```python
date_list=[]
for i in range(0,584):
    dates=extract_dates(fdata["feed__data__message"].iloc[i])
    date_list.append(dates)
#date_list
```

```python
for i in range(0,584):
    fdata["feed__data__updated_time"].iloc[i]=date_list[i]
```

```python
fdata["Location_list"]=fdata["feed__data__message"].str.find("vestalis gracilis")
fdata.head()
```

## Location

```python
qwerty=[]
for i in range(0,584):
    places = geograpy3.get_place_context(text=fdata["feed__data__message"].iloc[i])
    # print(fdata["feed__data__message"].iloc[i])
    x=places.regions
    qwerty.append(" ".join(x))
    print("completed",i,"/584")
    #print(places.regions)
#qwerty
```

## Remove Species name and Common name from Locatiob Data

```python
species1=pd.read_csv("Dragonfly_species.csv")
#species1["Aciagrion"].str.split()
cname1=pd.read_csv("cname.csv")
```

```python
cname1["COMMON NAME"]=cname1["COMMON NAME"].str.split()
cname1.head()
```

## Create a list of all Common Names

```python
names=[]
for i in range(0,108):
    for j in cname1["COMMON NAME"].iloc[i]:
        names.append(j)
len(names)
```

## Remove Irrelevant Location Data

```python
for i in range(0,584):
    for j in range(0,699):
        if len(qwerty[i]) != 0:
            qwerty[i]=qwerty[i].replace(species1["Aciagrion"].iloc[j],' ')

for i in range(0,584):
    for j in range(0,233):
        if len(qwerty[i]) != 0:
            qwerty[i]=qwerty[i].replace(names[j],' ')      
fdata["Location_list"]=qwerty
```

## Add Places

```python
places=pd.read_csv("Places.csv")
for i in range(0,places.shape[0]):
    s=places["PLACE"].iloc[i]
    result=''.join([i for i in s if not i.isdigit()])
    places["PLACE"].iloc[i]=result
places["PLACE"]=places["PLACE"].str.split(",")
allplaces=[]
for i in range(0,places.shape[0]):
    for j in places["PLACE"].iloc[i]:
        allplaces.append(j)
len(allplaces)
```

## Add Districts

```python
ldata=pd.read_csv("data_600.csv")
```

```python
dataw=ldata.iloc[:,6:]
listt3=dataw.columns
#dataw.head()
```

```python
fdata["Other matches for location"]=fdata["feed__data__message"].str.find("vestalis gracilis")
for i in range(0,584):
    for j in listt3:
        if dataw[j].iloc[i] >= 0:
            fdata["Other matches for location"].iloc[i]=j
```
